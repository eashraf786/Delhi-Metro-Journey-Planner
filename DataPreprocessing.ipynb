```python
import pandas as pd
df1 = pd.read_excel("D:/MTECH CSE/Adv Algo Lab Project/Delhi metro.xlsx",sheet_name='Delhi metro (2)')
df1 = df1.iloc[:,:4]
df1.head()
```

```python
stcl = dict(zip(df1['Station Names'],df1['Metro Line']))
len(stcl)
```

```python
import networkx as nx
g = nx.Graph()
for i in range(df1.shape[0]):
    if i!=df1.shape[0]-1 and df1.iloc[i,3] == df1.iloc[i+1,3]:
        wt = df1.iloc[i+1,2] - df1.iloc[i,2]
        g.add_edge(df1.iloc[i,1],df1.iloc[i+1,1],weight=wt)
```

```python
deg = dict(g.degree)
sdeg = dict(sorted(deg.items(), key=lambda item: item[1]))
sdeg
```

```python
pd.Series(sdeg.values()).value_counts()
```

```python
adj = nx.adjacency_matrix(g,nodelist=list(deg.keys())).toarray()
adj
```

```python
stat = list(deg.keys())
```

```python
pd.Series(stcl.values()).value_counts()
```

```python
for i in stat:
    print(f"{i},{stcl[i]}")
```

```python
adjdf = pd.DataFrame(adj,columns=stat,index=stat)
adjdf.to_csv("D:/MTECH CSE/Adv Algo Lab Project/DMRC_Adj_Matrix.csv",header=False,index=False)
```
